#### 2.1. Изучите распределение домов от наличия вида на набережную

## Домашнее задание 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
<br>Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df[['price', 'bedrooms', 'bathrooms', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement']].corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=0.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

__Вывод: Чем выше качество конструкции и дизайн, тем больше стоимость дома__

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg');
```

__Вывод: Наибольшее количество домов, которые находятся выше земли, имеют небольшую стоимость__

```python
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='kde');
```

__Вывод: Наибольшее количество домов имеют количество ванных комнат от 1 до 3, и при этом их стоимость невысокая__

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['price'], y=df['view'].astype('str'), showfliers = False)
plt.xlabel('price')
plt.ylabel('view')
plt.title('Зависимость стоимости домов от вида');
```

__Вывод: Наивысшую стоимость домов имеет оценку вида "4"__

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['price'], y=df['condition'].astype('str'), whis=5)
plt.xlabel('price')
plt.ylabel('condition')
plt.title('Зависимость стоимости домов от его состояния');
```

__Вывод: Чем выше оценка состояния дома, тем выше его стоимость, но при этом наивысшие стоимости домов имеют оценку состояния "4" и "3"__
